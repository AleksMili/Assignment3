```python

# question1  
import pandas as pd

# Import the dataset from the provided address
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"
users = pd.read_csv(url, sep='|')

# Discover the mean age per occupation
mean_age_per_occupation = users.groupby('occupation')['age'].mean()
print("Mean Age per Occupation:")
print(mean_age_per_occupation)

#  Discover the Male ratio per occupation and sort it
def male_ratio(group):
    return (group == 'M').sum() / group.size

male_ratio_per_occupation = users.groupby('occupation')['gender'].apply(male_ratio)
sorted_male_ratio = male_ratio_per_occupation.sort_values(ascending=False)
print("\nMale Ratio per Occupation (sorted):")
print(sorted_male_ratio)

# For each occupation, calculate the minimum and maximum ages
age_range_per_occupation = users.groupby('occupation')['age'].agg(['min', 'max'])
print("\nAge Range per Occupation:")
print(age_range_per_occupation)

# For each combination of occupation and sex, calculate the mean age
mean_age_per_combination = users.groupby(['occupation', 'gender'])['age'].mean()
print("\nMean Age per Combination of Occupation and Gender:")
print(mean_age_per_combination)

# For each occupation, present the percentage of women and men
def gender_percentage(group):
    total_count = group.size
    return (group.value_counts() / total_count) * 100

gender_percentage_per_occupation = users.groupby('occupation')['gender'].apply(gender_percentage)
print("\nGender Percentage per Occupation:")
print(gender_percentage_per_occupation)
```

```python
# question2 
import pandas as pd

# Import the dataset from the provided address
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv"
euro12 = pd.read_csv(url)

#  Assign it to a variable called euro12

#  Select only the Goal column
goals_column = euro12['Goals']
print("Goal column:")
print(goals_column)

#  How many teams participated in Euro2012?
num_teams = euro12['Team'].nunique()
print("\nNumber of teams participated in Euro2012:", num_teams)

#  What is the number of columns in the dataset?
num_columns = len(euro12.columns)
print("\nNumber of columns in the dataset:", num_columns)

# View only the columns Team, Yellow Cards, and Red Cards and assign them to a dataframe called discipline
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print("\nDiscipline dataframe:")
print(discipline)

#  Sort the teams by Red Cards, then by Yellow Cards
discipline_sorted = discipline.sort_values(by=['Red Cards', 'Yellow Cards'], ascending=False)
print("\nTeams sorted by Red Cards and then by Yellow Cards:")
print(discipline_sorted)

#  Calculate the mean Yellow Cards given per Team
mean_yellow_cards = discipline['Yellow Cards'].mean()
print("\nMean Yellow Cards given per Team:", mean_yellow_cards)

#  Filter teams that scored more than 6 goals
teams_more_than_6_goals = euro12[euro12['Goals'] > 6]
print("\nTeams that scored more than 6 goals:")
print(teams_more_than_6_goals)

#  Select the teams that start with G
teams_starting_with_G = euro12[euro12['Team'].str.startswith('G')]
print("\nTeams that start with G:")
print(teams_starting_with_G)

# Select the first 7 columns
first_7_columns = euro12.iloc[:, :7]
print("\nFirst 7 columns:")
print(first_7_columns)

#  Select all columns except the last 3
all_columns_except_last_3 = euro12.iloc[:, :-3]
print("\nAll columns except the last 3:")
print(all_columns_except_last_3)

# Present only the Shooting Accuracy from England, Italy, and Russia
teams_of_interest = ['England', 'Italy', 'Russia']
shooting_accuracy_selected_teams = euro12[euro12['Team'].isin(teams_of_interest)][['Team', 'Shooting Accuracy']]
print("\nShooting Accuracy from England, Italy, and Russia:")
print(shooting_accuracy_selected_teams)
```

```python
# question 4 
import pandas as pd
import numpy as np

#  Import the dataset
file_path = "C:/Users/acomi/Downloads/wind.txt"
data = pd.read_csv(file_path, delim_whitespace=True, parse_dates=[[0, 1, 2]], keep_date_col=True, index_col=0)
data.columns = data.columns.str.replace("VAL", "VAL1")  # Correcting a typo in the column name

#  Fix year 2061
def fix_year(year):
    if year > 2000:
        year -= 100
    return year

data.index = data.index.map(lambda x: pd.Timestamp(year=fix_year(x.year), month=x.month, day=x.day))

#  Count missing values for each location
missing_count = data.isnull().sum()

# Count non-missing values
non_missing_count = data.notnull().sum().sum()

#  Calculate the mean windspeeds
mean_windspeed = data.mean().mean()

#  Calculate location statistics
loc_stats = pd.DataFrame({
    'min': data.min(),
    'max': data.max(),
    'mean': data.mean(),
    'std': data.std()
})

#  Calculate daily statistics
day_stats = pd.DataFrame({
    'min': data.min(axis=1),
    'max': data.max(axis=1),
    'mean': data.mean(axis=1),
    'std': data.std(axis=1)
})

#  Average windspeed in January for each location
jan_avg = data[data.index.month == 1].mean()

#  Downsample to yearly frequency
yearly_data = data.resample('Y').mean()

#  Downsample to monthly frequency
monthly_data = data.resample('M').mean()

#  Downsample to weekly frequency
weekly_data = data.resample('W-MON').mean()  # Assuming weeks start on Monday

#  Calculate weekly statistics for the first 52 weeks
weekly_stats = weekly_data.head(52).agg(['min', 'max', 'mean', 'std'])

# Print results or perform further analysis as needed
print("Missing values count:\n", missing_count)
print("\nNon-missing values count:", non_missing_count)
print("\nMean windspeed:", mean_windspeed)
print("\nLocation statistics:\n", loc_stats)
print("\nDay statistics:\n", day_stats)
print("\nAverage windspeed in January:\n", jan_avg)
print("\nYearly data:\n", yearly_data)
print("\nMonthly data:\n", monthly_data)
print("\nWeekly data:\n", weekly_data)
print("\nWeekly statistics for the first 52 weeks:\n", weekly_stats)


```

```python
# question 5 
import pandas as pd

 Import the dataset from the provided address
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep='\t')

#  Assign it to a variable called chipo

#  See the first 10 entries
print("First 10 entries:")
print(chipo.head(10))

#  What is the number of observations in the dataset?
num_observations = chipo.shape[0]
print("\nNumber of observations:", num_observations)

#  What is the number of columns in the dataset?
num_columns = chipo.shape[1]
print("\nNumber of columns:", num_columns)

#  Print the name of all the columns.
print("\nColumn names:")
print(chipo.columns)

#  How is the dataset indexed?
print("\nIndex:")
print(chipo.index)

#  Which was the most-ordered item?
most_ordered_item = chipo['item_name'].value_counts().idxmax()
print("\nMost-ordered item:", most_ordered_item)

#  For the most-ordered item, how many items were ordered?
most_ordered_item_quantity = chipo[chipo['item_name'] == most_ordered_item]['quantity'].sum()
print("\nQuantity of most-ordered item:", most_ordered_item_quantity)

#  What was the most ordered item in the choice_description column?
most_ordered_choice_description = chipo['choice_description'].value_counts().idxmax()
print("\nMost ordered item in choice_description:", most_ordered_choice_description)

#  How many items were ordered in total?
total_items_ordered = chipo['quantity'].sum()
print("\nTotal items ordered:", total_items_ordered)


# Turn the item price into a float
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))

# Check the item price type
print("\nType of item_price after conversion:", chipo['item_price'].dtype)

#  How much was the revenue for the period in the dataset?
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("\nTotal revenue:", revenue)

#  How many orders were made in the period?
num_orders = chipo['order_id'].nunique()
print("\nNumber of orders:", num_orders)

#  What is the average revenue amount per order?
average_revenue_per_order = revenue / num_orders
print("\nAverage revenue per order:", average_revenue_per_order)

#  How many different items are sold?
num_different_items = chipo['item_name'].nunique()
print("\nNumber of different items sold:", num_different_items)
```

```python
#question 7

import pandas as pd
import matplotlib.pyplot as plt

# Specify the absolute path to the CSV file
file_path = "C:/Users/acomi/Downloads/us-marriages-divorces-1867-2014.csv"

# Load the CSV data into a DataFrame
data = pd.read_csv(file_path)

# Filter data for the years 1900, 1950, and 2000
years_to_plot = [1900, 1950, 2000]
filtered_data = data[data['Year'].isin(years_to_plot)]

# Extract the required columns
years = filtered_data['Year']
marriages_per_capita = filtered_data['Marriages_per_1000']
divorces_per_capita = filtered_data['Divorces_per_1000']

# Create a vertical bar chart
plt.figure(figsize=(10, 6))
bar_width = 0.35
plt.bar(years - bar_width/2, marriages_per_capita, bar_width, label='Marriages per 1000')
plt.bar(years + bar_width/2, divorces_per_capita, bar_width, label='Divorces per 1000')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Per Capita')
plt.title('Comparison of Marriages and Divorces per Capita in the U.S. (1900, 1950, 2000)')
plt.xticks(years)
plt.legend()

# Show the plot
plt.show()
```

```python

#question 8

import pandas as pd
import matplotlib.pyplot as plt

file_path = "C:/Users/acomi/Downloads/actor_kill_counts.csv"

data = pd.read_csv(file_path)

# Sort the data by kill count in descending order
sorted_data = data.sort_values(by='Count', ascending=True)

# Extract the actor names and kill counts
actor_names = sorted_data['Actor']
kill_counts = sorted_data['Count']

# Create a horizontal bar chart
plt.figure(figsize=(10, 8))
plt.barh(actor_names, kill_counts, color='skyblue')
plt.xlabel('Kill Count')
plt.ylabel('Actor')
plt.title('Deadliest Actors in Hollywood')
plt.tight_layout()

# Add labels to each bar
for i, count in enumerate(kill_counts):
    plt.text(count + 1, i, str(count), va='center')

# Show the plot
plt.show()
```

```python
#question 9

import pandas as pd
import matplotlib.pyplot as plt

# Specify the absolute path to the CSV file
file_path = "C:/Users/acomi/Downloads/roman-emperor-reigns.csv"

# Load the CSV data into a DataFrame
data = pd.read_csv(file_path)

# Count the occurrences of 'Assassinated' and 'Natural Causes'
assassinated_count = data[data['Cause_of_Death'] == 'Assassinated'].shape[0]
natural_causes_count = data[data['Cause_of_Death'] == 'Natural Causes'].shape[0]

# Calculate the percentages
total_emperors = assassinated_count + natural_causes_count
assassinated_percentage = (assassinated_count / total_emperors) * 100
natural_causes_percentage = (natural_causes_count / total_emperors) * 100

# Create a pie chart
labels = ['Assassinated', 'Natural Causes']
sizes = [assassinated_percentage, natural_causes_percentage]
colors = ['lightcoral', 'lightskyblue']
explode = (0.1, 0)  # Explode the 'Assassinated' slice

plt.figure(figsize=(6, 6))  # To ensure an even circle
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.title('Fraction of Roman Emperors Assassinated')
plt.tight_layout()

# Show the pie chart
plt.show()

```
